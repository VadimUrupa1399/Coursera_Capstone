```python
import pandas as pd
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

wikitable = pd.read_html(url, attrs={'class': 'wikitable sortable'})
print('Extracted {num} vikitables:'.format(num=(len(wikitable))))

#import pandas and scrape 1 table
```

```python
wikitable[0].head(5)
```

```python
data_table = wikitable[0]
data_table
#
```

```python
data_table.to_csv('data_table.csv')
```
